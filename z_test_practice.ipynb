# Z-Test Practice:
This notebook demonstrates T-Test for hypothesis testing:
- One-sample Z-test (one-tailed & two-tailed)
- Visualizations to understand rejection regions and group differences


### **ONE TAILED REJECTION Z-TEST LAYOUT**
EXAMPLE 1:A teacher claims the average exam score = 70. We take a sample of 50 students, and the sample mean = 74,with a population standard deviation = 10.
#We want to check if students really scored higher than 70.


```python


import numpy as np
from scipy import stats
from IPython.display import display, HTML

pop_mean = 70
sample_mean = 74
std_dev = 10
n = 50
# z statistics
z=(sample_mean - pop_mean)/(std_dev/np.sqrt(n))
p_value= 2*(1-stats.norm.cdf(z))
display(HTML(f"<b style='color:blue;'>z={z}</b>"))
display(HTML(f"<b style='color:blue;'>p_value={p_value}</b>"))
#Interpretation
if p_value < 0.05:
    display(HTML("<b style='color:green;'>✅ We reject the null hypothesis: students scored significantly higher than 70.</b>"))
else:
    display(HTML("<b style='color:red;'>❌ We fail to reject the null hypothesis: no significant evidence that students scored higher than 70.</b>"))

#VISUALIZATION
import matplotlib.pyplot as plt
x=np.linspace(-4,4,1000)
y=stats.norm.pdf(x,0,1)
plt.plot(x,y,color='black',lw=2)
x_reject=np.linspace(1.645,4,300)
y_reject=stats.norm.pdf(x_reject,0,1)
plt.fill_between(x_reject,y_reject,color='red',alpha=0.3, label='Rejection Region')
plt.legend()
plt.axvline(z,color='blue',linestyle='--',label=f'z-value:{z:.2f}')
plt.text(z+0.1,0.05,f'Observed Z = {z:.2f}',color='blue')
plt.text(2.1,0.15,'Reject H0 --> Higher than 70',color='red',fontsize=10)
plt.title("Z-test :Sampling Distribution")
plt.xlabel("Z-score")
plt.ylabel("Probability Density")
plt.legend()
plt.grid(alpha=0.4)
plt.show()
```

###**TWO TAILED REJECTION LAYOUT**
Check if students scored differently from 70 (could be higher or lower). Two-tailed test.




```python
import numpy as np
from scipy import stats
import matplotlib.pyplot as plt
from IPython.display import display, HTML

# Given values
pop_mean = 70
sample_mean = 74
std_dev = 10
n = 50

# Z-statistic
z = (sample_mean - pop_mean) / (std_dev / np.sqrt(n))

# Two-tailed p-value
p_value = 2 * (1 - stats.norm.cdf(abs(z)))

# Display results
display(HTML(f"<b style='color:blue;'>Z = {z:.2f}</b>"))
display(HTML(f"<b style='color:blue;'>P-value = {p_value:.4f}</b>"))

# Interpretation
if p_value < 0.05:
    display(HTML("<b style='color:green;'>✅ We reject H₀: Students' average score is significantly different from 70.</b>"))
else:
    display(HTML("<b style='color:red;'>❌ We fail to reject H₀: No significant difference from 70.</b>"))

# Visualization
x = np.linspace(-4, 4, 1000)
y = stats.norm.pdf(x, 0, 1)

plt.plot(x, y, color='black', lw=2)

# Shade rejection regions (both tails)
x_left = np.linspace(-1.96, -4, 300)
x_right = np.linspace(1.96, 4, 300)
plt.fill_between(x_left, stats.norm.pdf(x_left, 0, 1), color='red', alpha=0.3, label='Left Rejection Region')
plt.fill_between(x_right, stats.norm.pdf(x_right, 0, 1), color='red', alpha=0.3, label='Right Rejection Region')

# Draw our observed z
plt.axvline(z, color='blue', linestyle='--', lw=2, label=f'Observed Z = {z:.2f}')

plt.title("Two-tailed Z-test: Sampling Distribution")
plt.xlabel("Z-Score")
plt.ylabel("Probability Density")
plt.legend()
plt.grid(alpha=0.3)
plt.show()

```
