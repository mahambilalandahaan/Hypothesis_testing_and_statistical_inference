### **CHI-SQUARE_TEST**

EXAMPLE 1: DOES GENDER EFFECT WHICH PRODUCT PEOPLE PREFER

```python
#EXAMPLE 1: DOES GENDER EFFECY WHICH PRODUCT PEOPLE PREFER
#Step 1: IMPORTS
import pandas as pd
from scipy.stats import chi2_contingency
import matplotlib.pyplot as plt

#STEP 2: CREATE CONTIGENCY TABLE
data =pd.DataFrame({
    'Gender':['Male','Female','Male','Female','Male','Female','Male'],
    'Product':['A','B','C','B','A','B','A']
})

#STEP 3: CREATE CROSS-TAB (CONTIGENCY TABLE)
table= pd.crosstab(data['Gender'],data['Product'])
print(f"CONTIGENCY TABLE :\n {table} \n")

#STEP4: PERFORM CHI-SQUARE TEST FOR INDEPENDENCE
chi2 , p , dof , expected = chi2_contingency(table)
print(f"CHI-SQUARE VALUE : {chi2:.3f}")
print(f"P-VALUE : {p:.2f}")
print(f"DEGREE OF FREEDOM : {dof}")
print(f"EXPECTED FREQUENCIES : \n {expected}")

#STEP5: INTERPRETATION
from IPython.display import display, HTML

if p < 0.05:
    display(HTML("<b style='color:green;'>There is a significant association between gender and product preference.</b>"))
else:
    display(HTML("<b style='color:red;'>There is no significant association between gender and product preference.</b>"))

#STEP 6: VISUALIZATION
# Step 4: Visualize
table.plot(kind='bar', stacked=False, figsize=(6,4), edgecolor='black')
plt.title(f'Product Preference by Gender (p :{p:.2f})')
plt.xlabel('Gender')
plt.ylabel('Count')
plt.legend(title='Product')
plt.grid(axis='y', linestyle='--')
plt.show()
```
